### Deriving columns from the data for further analysis

```python
df['issue_d'] = pd.to_datetime(df['issue_d'], format='%b-%y')
df['issue_d_year']=df['issue_d'].dt.year
df['issue_d_month']=df['issue_d'].dt.month
df['total_no_loans']= df.groupby('issue_d_year')['loan_status'].count()
df['annual_inc_range']= pd.cut(df['annual_inc'],[0,30000, 60000, 90000, 120000, 150000], include_lowest=True, labels =['0-30k','30k-60k','60k-90k','91k-120k','121k+'])
df['int_rate_range']=pd.cut(df['int_rate'], [0, 4, 8, 12, 16, 20], labels=['0-4%','4-8%', '8-12%', '12-16%', '16%+'])
df['dti_range'] = pd.cut(df['dti'], [0, 5, 10, 15, 20, 25, 30, 35], labels=['0-5','5-10', '10-15', '15-20', '20-25', '25-30', '30+'])
df['loan_amount_range'] = pd.cut(df['loan_amnt'], [0, 5000, 10000, 15000, 20000, 25000, 30000, 35000], labels=['0-5000','5k-10k', '10k-15k', '15k-20k', '20k-25k', '25k-30k', '30+'])
```

##### Defining a function to Calculate Default Ratio- We can call this function in the Bivariate analysis.

```python
def defaultratio(df,column):
    inc = df.groupby([column, 'loan_status']).loan_status.count().unstack().fillna(0).reset_index()
    inc['Total'] = inc['Charged Off'] + inc['Fully Paid'] 
    inc['Defaults_Ratio'] = inc['Charged Off'] / inc['Total']
    inc.sort_values('Defaults_Ratio', ascending=False)
    return inc
```
